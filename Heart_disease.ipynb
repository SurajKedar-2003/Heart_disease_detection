```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("heart_deasease.csv")
```

```python
plt.figure(figsize=(20,20))
ax = sns.boxplot(data=df)
```

```python
from scipy import stats
z = np.abs(stats.zscore(df))
print(z)
```

```python
threshold = 3
print(np.where(z>3))
```

```python
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 -Q1
print(IQR)
```

```python
df = df[(z<3).all(axis=1)]
df.shape
```

```python
df = df[~((df < (Q1 - 1.5*IQR)) | (df>(Q3 + 1.5*IQR))).any(axis=1)]
df.shape
```

```python
plt.figure(figsize = (20, 20))
ax = sns.boxplot(data = df)
```

```python
plt.figure(figsize = (20, 20))
d = sns.heatmap(df.corr(), cmap = "coolwarm", annot = True)
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
dataset = pd.get_dummies(df, columns = ['sex', 'cp', 'fbs', 'restecg', 'exang', 'ca', 'thal'])
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
dataset[columns_to_scale] = standardScaler.fit_transform(dataset[columns_to_scale])
```

```python
dataset.head()
```

```python
dataset.tail()
```

```python
dataset.describe()
```

```python
sns.pairplot(df, hue="target", height = 3, aspect =1);
```

```python
#model selection
y = dataset['target']
X = dataset.drop(['target'], axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 5)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
#KNN classifier
knn_classifier = KNeighborsClassifier(n_neighbors = 5)
knn_classifier.fit(X_train, y_train)
score = cross_val_score(knn_classifier, X_train, y_train, cv = 10)
y_pred_knn = knn_classifier.predict(X_test)
accuracy_score(y_test, y_pred_knn)
```

```python
score.mean()
```

```python
knn_Classifier = KNeighborsClassifier(algorithm='auto', leaf_size = 30, metric = 'minkowski', metric_params = None, n_jobs =1, n_neighbors = 5,p=1, weights = 'uniform')
knn_Classifier.fit(X_train, y_train)
score = cross_val_score(knn_classifier, X_train, y_train, cv=10)
y_pred_knn = knn_classifier.predict(X_test)
accuracy_score(y_test, y_pred_knn)
```

```python
score=cross_val_score(knn_classifier,X_train,y_train,cv=10)
score.mean()
```

```python
#confustion metrix
cm = confusion_matrix(y_test, y_pred_knn)
plt.title('Heatmap of Confusion Matrix', fontsize = 15)
sns.heatmap(cm, annot = True)
plt.show()
```

```python
print(classification_report(y_test, y_pred_knn))
```

```python
#random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf_classifier = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf_classifier.fit(X_train, y_train)
y_pred_rf = rf_classifier.predict(X_test)
accuracy_score(y_test, y_pred_rf)
```

```python
rf_cm = confusion_matrix(y_test, y_pred_rf)
plt.title('Heatmap of Confusion Matrix', fontsize = 15)
sns.heatmap(cm, annot = True)
plt.show()
```

```python
score=cross_val_score(rf_classifier,X_train,y_train,cv=10)
score.mean()
```
